## Project Title
Quantra Retail Performance and Business Intelligence Report

## Background
Quantra Retail is a fictional UK-based retail company operating across 10 major cities including London, Manchester, Birmingham, Leeds, Bristol, Edinburgh, Glasgow, Liverpool, Sheffield, and Cardiff. The company offers products across 6 key categories. Electronics, Clothing, Groceries, Home & Garden, Beauty, and Sports through both online and in-store channels. 

As the business scales across regions, understanding sales performance, customer behaviour, and product profitability becomes critical for driving revenue growth and operational efficiency.


## Business Problem
Quantra Retail operates across multiple cities and product categories 
but lacks a unified view of its commercial performance. The leadership 
team cannot clearly identify which regions are underperforming, which 
product categories are most profitable, or how customer purchasing 
behaviour varies across demographics. Without these insights, resource 
allocation, promotional strategies, and inventory decisions are being 
made without data backing — creating a significant risk of lost revenue 
and missed growth opportunities.

## Project Aim
To transform Quantra Retail's raw transactional data into a clean, 
structured dataset that powers meaningful SQL analysis and an 
interactive Power BI dashboard — enabling the business to make 
faster, smarter, and more confident commercial decisions.

1. Clean and validate all datasets to ensure accuracy and consistency
2. Identify top performing product categories and regions by revenue
3. Understand customer demographics and purchasing behaviour
4. Uncover seasonal and monthly sales trends across 2024–2026
5. Compare online vs in-store channel performance
6. Segment customers by value to support targeted marketing strategies
7. Deliver actionable recommendations backed by data evidence

#### Dataset
- `products.csv` → 30 products across 6 categories (product_id, product_name, product category, unit_price)
- `customers.csv` → 1,000 UK customers (customer_id, customer name, city, age, gender, email, join date segment)
- `transactions.csv` → 10,000 retail transactions (transaction_id, customer_id, product_id, date, quantity, unit price, store type, payment method, total amount)

## Data Cleaning 
Fix date columns in customers and transactions tables

```python
customers_df['join_date'] = pd.to_datetime(customers_df['join_date'])
transactions_df['date'] = pd.to_datetime(transactions_df['date'])
```

##### Extract useful date features from transactions

```python
transactions_df['year'] = transactions_df['date'].dt.year
transactions_df['month'] = transactions_df['date'].dt.month
transactions_df['month_name'] = transactions_df['date'].dt.strftime('%B')
transactions_df['day_of_week'] = transactions_df['date'].dt.strftime('%A')
```

##### Verify

```python
print("Customers join_date dtype:", customers_df['join_date'].dtype)
print("Transactions date dtype:", transactions_df['date'].dtype)
print("\n✅ Date columns fixed successfully!")
transactions_df.head()
```

##### Adding profit column

```python
transactions_df = transactions_df.merge(
    products_df[['product_id', 'cost_price']], 
    on='product_id', how='left')

transactions_df['profit'] = round(
    transactions_df['total_amount'] - 
    (transactions_df['cost_price'] * transactions_df['quantity']), 2)

print("✅ Profit column added!")
transactions_df.head()
```

```python
transactions_df['profit'] = round(
    transactions_df['total_amount'] - 
    (transactions_df['cost_price'] * transactions_df['quantity']), 2)

transactions_df['profit_margin'] = round(
    (transactions_df['profit'] / transactions_df['total_amount']) * 100, 2)
```

```python
print(" Profit and Profit Margin columns added!")
print(f"\nAverage profit margin: {transactions_df['profit_margin'].mean().round(2)}%")
transactions_df[['transaction_id', 'total_amount', 'cost_price', 'profit', 'profit_margin']].head()
```

#### Profit Margin Analysis 

```python
print("=" * 40)
print("PROFIT MARGIN ANALYSIS")
print("=" * 40)
print(f"Average profit margin: {transactions_df['profit_margin'].mean().round(2)}%")
print(f"Highest profit margin: {transactions_df['profit_margin'].max().round(2)}%")
print(f"Lowest profit margin: {transactions_df['profit_margin'].min().round(2)}%")
print(f"\nTotal Revenue: £{transactions_df['total_amount'].sum().round(2):,}")
print(f"Total Cost: £{(transactions_df['cost_price'] * transactions_df['quantity']).sum().round(2):,}")
print(f"Total Profit: £{transactions_df['profit'].sum().round(2):,}")
```

#### Profit margin insight
- Quantra Retail maintains a healthy average profit margin of **47.75%**
- This means for every £1 of revenue generated, approximately 
**48 pence is profit**
- Total revenue of **£7.4M** generated **£3.5M** in profit 
across 10,000 transactions
- Margin varies across transactions due to differences in product 
categories and cost structures
- Further analysis in SQL and Power BI will reveal which product 
categories and regions are most profitable
