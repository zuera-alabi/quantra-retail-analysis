#### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from faker import Faker
import random
from datetime import datetime, timedelta
```

#### Set faker object to UK English, set random seed to 42 so it give identical data for every run

```python
fake = Faker('en_GB')
np.random.seed(42)
random.seed(42)

print("Libraries imported successfully!")
```

#### Generate product table 

```python
categories = {
    'Electronics': ['Laptop', 'Smartphone', 'Tablet', 'Headphones', 'Smart Watch'],
    'Clothing': ['Jacket', 'Jeans', 'Dress', 'Trainers', 'Coat'],
    'Groceries': ['Coffee', 'Tea', 'Olive Oil', 'Pasta', 'Cereal'],
    'Home & Garden': ['Sofa', 'Lamp', 'Plant Pot', 'Cushion', 'Rug'],
    'Beauty': ['Moisturiser', 'Shampoo', 'Perfume', 'Lipstick', 'Foundation'],
    'Sports': ['Yoga Mat', 'Dumbbells', 'Running Shoes', 'Protein Powder', 'Gym Bag']
}
products = []
product_id = 1

for category, items in categories.items():
    for item in items:
        price = round(random.uniform(5, 500), 2)
        products.append({
            'product_id': f'P{product_id:03d}',
            'product_name': item,
            'product_category': category,
            'unit_price': price,
            'cost_price': round(price * random.uniform(0.4, 0.7), 2)
        })
        product_id += 1

        products_df = pd.DataFrame(products)
print(products_df.head(10))
print(f"\nTotal products: {len(products_df)}")
```

#### Generate customers table 

```python
uk_cities = ['London', 'Manchester', 'Birmingham', 'Leeds', 'Bristol', 
 'Edinburgh', 'Glasgow', 'Liverpool', 'Sheffield', 'Cardiff']

customers = []
for i in range(1, 1001):
    customers.append({
        'customer_id': f'C{i:04d}',
        'customer_name': fake.name(),
        'city': random.choice(uk_cities),
        'age': random.randint(18, 70),
        'gender': random.choice(['Male', 'Female']),
        'email': fake.email(),
        'join_date': fake.date_between(start_date='-3y', end_date='today')
    })

customers_df = pd.DataFrame(customers)
print(customers_df.head(10))
print(f"\nTotal customers: {len(customers_df)}")
```

```python
customers_df.shape
```

#### Generate transactions table 

```python
transactions = []

for i in range(1, 10001):
    customer = random.choice(customers_df['customer_id'].tolist())
    product = products_df.sample(1).iloc[0]
    quantity = random.randint(1, 5)
    
    transactions.append({
        'transaction_id': f'T{i:05d}',
        'customer_id': customer,
        'product_id': product['product_id'],
        'quantity': quantity,
        'unit_price': product['unit_price'],
        'total_amount': round(quantity * product['unit_price'], 2),
        'date': fake.date_between(start_date='-2y', end_date='today'),
        'store_type': random.choice(['Online', 'In-Store']),
        'payment_method': random.choice(['Credit Card', 'Debit Card', 'PayPal', 'Apple Pay', 'Cash'])
    })

transactions_df = pd.DataFrame(transactions)
print(transactions_df.head(10))
print(f"\nTotal transactions: {len(transactions_df)}")
```
